Для каждой из частей выгружаем детали по заведениям по id через цикл

```python
# Пустой датафрейм для хранения данных о заведениях
yelp_rest_details = pd.DataFrame()
```

```python
# Цикл по каждому ресторану из столбца id
for id in yelp_rest_bars_1['id']:
    url = "https://red-flower-business-data.p.rapidapi.com/business-details"

    querystring = {"business_id":str(id),"yelp_domain":"yelp.co.uk"}

    headers = {
        #"x-rapidapi-key": "057acdba41msh5565e18a973aa7ap138573jsn094c698cf753",
        #"x-rapidapi-host": "red-flower-business-data.p.rapidapi.com"
    }

    # Отправляем запрос и получаем данные
    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code == 200:
        # Преобразуем ответ в JSON
        info = response.json()

        # Преобразуем данные ресторана в датафрейм и добавляем его к основному
        temp_df = pd.DataFrame(info['data'])
        yelp_rest_details = pd.concat([yelp_rest_details, temp_df], ignore_index=True)
        print(f"Добавлены данные по заведению {id}")
    else:
        print(f"Ошибка при запросе ID {id}: {response.status_code}")
    time.sleep(1)

yelp_rest_details
```

Исправление ошибки, допущенной в цикле (код цикла в предыдущей ячейке обновлен)

```python
# "Расплющивание" списка словарей в отдельные строки
yelp_rest_details_upd = yelp_rest_details.explode('data')

# Преобразование каждого словаря в столбцы
yelp_rest_details_upd2 = pd.json_normalize(yelp_rest_details_upd['data'])

# Соединение с оригинальным датафреймом (если есть другие столбцы)
yelp_rest_details_res = yelp_rest_details_upd.drop(columns=['data']).reset_index(drop=True)
yelp_rest_details_res = pd.concat([yelp_rest_details_res, yelp_rest_details_upd2], axis=1)

yelp_rest_details_res = yelp_rest_details_res.drop(columns=['status', 'request_id']).reset_index(drop=True)
yelp_rest_details_res
```
