One-Hot encoding на основе объединения по типам кухни

```python
rest_cuisines = rest.copy()

# Разделяем строки в колонке 'cuisines' на отдельные кухни
rest_cuisines['cuisines'] = rest_cuisines['cuisines'].str.split(', ')

# Создаем список всех уникальных кухонь с учетом правил объединения
unique_cuisines = set()
for sublist in rest_cuisines['cuisines']:
    for cuisine in sublist:
        # Проверяем, принадлежит ли кухня к группе для объединения
        merged = False
        for new_cuisine, to_merge in merge_rules.items():
            if cuisine in to_merge:
                unique_cuisines.add(new_cuisine)
                merged = True
                break
        if not merged:
            unique_cuisines.add(cuisine)

# Создаем DataFrame с One-hot encoding
cuisines_dummies = pd.DataFrame({
    cuisine: rest_cuisines['cuisines'].apply(
        lambda x: 1 if any(c in x for c in (merge_rules.get(cuisine, []) + [cuisine] if cuisine in merge_rules else [cuisine])) else 0
    ) for cuisine in unique_cuisines
})

# Объединяем результат с исходным DataFrame
rest_cuisines = pd.concat([rest_cuisines, cuisines_dummies], axis=1)

rest_cuisines
```

**Доли ресторанов с рейтингом выше 4.8 по видам кухни**

```python
# Считаем долю ресторанов с рейтингом выше 4.8 для каждой категории
high_rating_cuisine = {}
for cuisine in cuisine_columns:
    # Фильтруем рестораны с рейтингом выше 4.8 и где категория = 1
    high_rating_count = rest_cuisines[(rest_cuisines['rating'] >= 4.8) & (rest_cuisines[cuisine] == 1)].shape[0]
    # Общее количество ресторанов в категории
    total_count = rest_cuisines[rest_cuisines[cuisine] == 1].shape[0]
    # Считаем долю
    if total_count > 0:
        high_rating_cuisine[cuisine] = high_rating_count / total_count
    else:
        high_rating_cuisine[cuisine] = 0  # Если в категории нет ресторанов

high_rating_cuisine_df = pd.DataFrame(list(high_rating_cuisine.items()), columns=['Cuisine', 'High Rating Cuisine'])

high_rating_cuisine_df
```

**Число отзывов по медиане с оценками 1-5 по виду кухни**

```python
# Создаем новый DataFrame для хранения данных по кухням
cuisine_reviews = []

# Проходим по каждой кухне
for cuisine in cuisine_columns:
    # Фильтруем рестораны, где кухня = 1
    filtered_df = rest_cuisines[rest_cuisines[cuisine] == 1]
    # Считаем медиану количества отзывов для этой кухни
    total_reviews = filtered_df['reviews'].median()
    # Считаем медиану количества отзывов с оценками 1-5 для этой кухни
    reviews_rating_1 = filtered_df['reviews_rating_1'].median()
    reviews_rating_2 = filtered_df['reviews_rating_2'].median()
    reviews_rating_3 = filtered_df['reviews_rating_3'].median()
    reviews_rating_4 = filtered_df['reviews_rating_4'].median()
    reviews_rating_5 = filtered_df['reviews_rating_5'].median()
    # Добавляем данные в список
    cuisine_reviews.append({
        'cuisine': cuisine,
        'total_reviews': total_reviews,
        'reviews_rating_1': reviews_rating_1,
        'reviews_rating_2': reviews_rating_2,
        'reviews_rating_3': reviews_rating_3,
        'reviews_rating_4': reviews_rating_4,
        'reviews_rating_5': reviews_rating_5
    })

cuisine_reviews_df = pd.DataFrame(cuisine_reviews)

stacked_data = []
for cuisine in cuisine_reviews_df['cuisine']:
    for i in range(1, 6):
        stacked_data.append({
            'cuisine': cuisine,
            'rating': f'Rating {i}',
            'count': cuisine_reviews_df.loc[cuisine_reviews_df['cuisine'] == cuisine, f'reviews_rating_{i}'].values[0]
        })

stacked_df = pd.DataFrame(stacked_data)

fig = px.bar(
    stacked_df,
    x='cuisine',
    y='count',
    color='rating'
)

fig.update_layout(
    title={
        'text': '<b>Число отзывов по медиане с оценками 1-5 по виду кухни</b>',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title='Кухня',
    yaxis_title='Число отзывов (медиана)',
    barmode='stack',
    width=800,
    height=600
)

fig.show()
```

One-Hot encoding для dining options

```python
rest_dinings = rest_diets.copy()

# Разделяем строки в колонке 'diets' на отдельные диеты
rest_dinings['dining_options'] = rest_diets['dining_options'].str.split(', ')

# Создаем список всех уникальных диет с учетом правил объединения
unique_dinings = set()
for sublist in rest_dinings['dining_options']:
    for dining in sublist:
        # Проверяем, принадлежит ли диета к группе для объединения
        merged = False
        for new_dining, to_merge in merge_rules_dinings.items():
            if dining in to_merge:
                unique_dinings.add(new_dining)
                merged = True
                break
        if not merged:
            unique_dinings.add(dining)

dinings_dummies = pd.DataFrame({
    dining: rest_dinings['dining_options'].apply(
        lambda x: 1 if any(c in x for c in (merge_rules_dinings.get(dining, []) + [dining] if dining in merge_rules_dinings else [dining])) else 0
    ) for dining in unique_dinings
})

rest_dinings = pd.concat([rest_dinings, dinings_dummies], axis=1)

rest_dinings
```
