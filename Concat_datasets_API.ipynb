# **Приводим данные в столбцах в нормальный вид**

Столбцы `cuisines`, `reviews_per_rating`, `diets`, `dining_options` содержат списки, словари и списки словарей.

```python
import re
```

```python
def process_columns(text):
  if isinstance(text, str):
    # Убираем скобки и кавычки
    text = text.replace("[", "").replace("]", "").replace("'", "").replace("{", "").replace("}", "")

    # Убираем url и оставляем только значение после title
    if "title:" in text:
        text = text.split("title: ")[1].split(",")[0]  # Извлечение значения после title до первой запятой

  return text
```

```python
copy_tripadvisor_yelp_data_non_dupl = tripadvisor_yelp_data_non_dupl.copy()

# Список столбцов, к которым нужно применить функцию
columns_to_process = ['cuisines', 'diets', 'dining_options']

# Применение функции к каждому столбцу в списке
for col in columns_to_process:
    copy_tripadvisor_yelp_data_non_dupl[col] = copy_tripadvisor_yelp_data_non_dupl[col].apply(process_columns)

copy_tripadvisor_yelp_data_non_dupl
```

Разделим столбец `reviews_per_rating` на 5 столбцов:

```python
import ast
```

```python
def extract_reviews_per_rating(row):
    if isinstance(row, str):  # Проверка, что это строка
        # Преобразуем строку в словарь
        rating_dict = ast.literal_eval(row)
        return pd.Series([rating_dict.get('1', None), rating_dict.get('2', None),
                          rating_dict.get('3', None), rating_dict.get('4', None),
                          rating_dict.get('5', None)])  # Возвращаем серию с 5 оценками
    else:
        return pd.Series([None, None, None, None, None])  # Для NaN возвращаем пустые значения
```

```python
rating_columns = copy_tripadvisor_yelp_data_non_dupl['reviews_per_rating'].apply(extract_reviews_per_rating)

# Переименовываем столбцы
rating_columns.columns = ['reviews_rating_1', 'reviews_rating_2', 'reviews_rating_3', 'reviews_rating_4', 'reviews_rating_5']

# Объединяем исходный датафрейм с новыми столбцами
copy_tripadvisor_yelp_data_non_dupl = pd.concat([copy_tripadvisor_yelp_data_non_dupl, rating_columns], axis=1)

# Убираем исходный столбец 'reviews'
tripadvisor_yelp_data_1 = copy_tripadvisor_yelp_data_non_dupl.drop(columns=['reviews_per_rating'])

tripadvisor_yelp_data_1
```

```python
tripadvisor_yelp_data_1.dtypes
```

Заполним пропуски в полученных столбцах на `-1` (`0` будет некорректно, т.к. информации нет)

Приведем значения по столбцам к типу `int`

```python
tripadvisor_yelp_data_ready = tripadvisor_yelp_data_1.copy()

cols_to_convert = ['reviews_rating_1', 'reviews_rating_2', 'reviews_rating_3', 'reviews_rating_4', 'reviews_rating_5']

tripadvisor_yelp_data_ready[cols_to_convert] = tripadvisor_yelp_data_ready[cols_to_convert].fillna(-1)
tripadvisor_yelp_data_ready[cols_to_convert] = tripadvisor_yelp_data_ready[cols_to_convert].astype(int)

tripadvisor_yelp_data_ready
```

Значения `''` и `None` заменим на `np.nan`

```python
import numpy as np
```

```python
tripadvisor_yelp_data_final = tripadvisor_yelp_data_ready.replace([None, ''], np.nan)
tripadvisor_yelp_data_final
```
